```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
import os     

os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
```

```python
# Set up all our data in a couple of data frames.
# Replace the path with the correct path for your data
df = pd.read_csv('ramen-ratings.csv')
```

```python
df.columns
```

```python
df=df.drop('Top Ten',axis=1)
```

```python
df['Stars']=df['Stars'].replace('Unrated', 0)
```

```python
df['Stars']=pd.to_numeric(df['Stars'])
```

```python
import math
df = df[df['Stars']!=0]
```

```python
df['Stars']=((df['Stars']-df['Stars'].min())/df['Stars'].max()-df['Stars'].min())
```

```python
df['Stars'].hist()
print(df['Stars'].mean())
```

```python
df['class']=np.where(df['Stars']>=df['Stars'].mean(),1,0)
```

```python
df
```

```python
import time
start_time = time.time()
# This is the model we'll be using.
from sklearn import tree

# A convenience for displaying visualizations.
from IPython.display import Image

# Packages for rendering our tree.
import pydotplus
import graphviz

X = df.drop(['Review #','Variety','Stars','class'], 1)
Y = df['class']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

# Initialize and train our tree.
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_depth=10
)
decision_tree.fit(X,Y)

# Render our tree.
dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=X.columns,
    class_names=['Not Good', 'Good'],
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
from sklearn.model_selection import cross_val_score
start_time = time.time()
cross_val_score(decision_tree, X, Y, cv=10)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
start_time = time.time()
rfc = ensemble.RandomForestClassifier()
X = df.drop(['Review #','Variety','Stars','class'], 1)
Y = df['class']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)
print("--- %s seconds ---" % (time.time() - start_time))
```
